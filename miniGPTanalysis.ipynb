```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt

def plot_comparison_of_val_losses(results1, results2, label1='Default MiniGPT', label2='Stack and Sum Method',
                                  title='Validation Loss Comparison', xlabel='Iteration', ylabel='Validation Loss'):
   
    # Extracting data for the first dataset
    iters1 = [d['iter'] for d in results1]
    val_losses1 = [d['val_loss'] for d in results1]
    val_losses2 = [d['val_loss'] for d in results2]
    #val_losses1 = [d['val_loss'] - s for d,s in zip(results1, results2)]

    plt.figure(figsize=(10, 6))
    
    plt.plot(iters1, val_losses1, '-o', color='royalblue', linewidth=2, markersize=5, label=label1)
    plt.plot(iters1, val_losses2, '-s', color='darkorange', linewidth=2, markersize=5, label=label2)
    
    plt.title(title, fontsize=16)
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel(ylabel, fontsize=14)
    
    plt.grid(True, which='both', linestyle='--', linewidth=0.5, color='grey')
    plt.legend(fontsize=12)
    # plt.tight_layout()
    plt.show()
```

```python
df = pd.read_csv('default_gpt.csv')
results_default = df.to_dict('records')
df_fun = pd.read_csv('fun_gpt.csv')
results_fun = df_fun.to_dict('records')
plot_comparison_of_val_losses(results_default, results_fun)
```

```python
import matplotlib.pyplot as plt

def plot_difference(results1, results2, label1='Default - Stack and Sum',
                                  title='Validation Loss Difference', xlabel='Iteration', ylabel='Difference'):
    # Extracting data for the first dataset
    iters1 = [d['iter'] for d in results1]
    val_losses1 = [d['val_loss'] - s['val_loss'] for d,s in zip(results1, results2)]

    plt.figure(figsize=(10, 6))
    plt.plot(iters1, val_losses1, '-o', color='royalblue', linewidth=2, markersize=5, label=label1)
    
    plt.title(title, fontsize=16)
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel(ylabel, fontsize=14)
    
    plt.grid(True, which='both', linestyle='--', linewidth=0.5, color='grey')
    plt.legend(fontsize=12)
    plt.tight_layout()
    plt.show()
```

```python
df = pd.read_csv('default_gpt.csv')
results_default = df.to_dict('records')
df_fun = pd.read_csv('fun_gpt.csv')
results_fun = df_fun.to_dict('records')
plot_difference(results_default, results_fun)
```
